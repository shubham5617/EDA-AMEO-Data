## Dataset Description:

- The dataset was released by Aspiring Minds from the Aspiring Mind Employment Outcome 2015 (AMEO). 
- The study is primarily limited  only to students with engineering disciplines. 
- The dataset contains the employment outcomes of engineering graduates as dependent variables (Salary, Job Titles, and Job Locations) along with the standardized scores from three different areas – cognitive skills, technical skills and personality skills. 
- The dataset also contains demographic features. 
- The dataset  contains  around  40 independent variables and 4000 data points. 
- The independent variables are both continuous and categorical in nature.

### Removing the Unnecessary features

```python
df = df.drop(columns= ['Unnamed: 0', 'ID', 'CollegeID', 'CollegeCityID', 'CollegeCityTier'])
```

### Datatype Conversion

```python
# Converting float to int datatype of salary feature
df['Salary'] = df['Salary'].astype('int')
```

#### As the survey was conducted back in 2015, we assume that the present value in DOL feature replace with the 31/12/2015 

```python
df['DOL'].replace('present', '2015-12-31', inplace=True)
```

```python
df['DOJ'] = pd.to_datetime(df['DOJ'])
df['DOL'] = pd.to_datetime(df['DOL'])
df['DOB'] = pd.to_datetime(df['DOB'])
```

```python
cat_feature = ['Gender', 'CollegeTier', 'Degree', 'Specialization', 'CollegeState']

for feat in cat_feature:
    df[feat] = df[feat].astype('category')
```

## Data Cleaning

### 1. Identify whether DOJ is greater than the DOL

```python
res = df[df['DOJ'] > df['DOL']].shape[0]
print('Date of leaving is earlier than the date of joining for {} datapoints'.format(res))
```

#### It looks like, there is some mistake while constructing the dataset so we remove these 40 datapoints from our dataset.

```python
df = df[df['DOJ'] < df['DOL']]
df.shape
```

### 2. Ensuring the results are in Percentage and not in a CGPA

```python
print((df['10percentage'] <= 10).sum())
print((df['12percentage'] <= 10).sum())
print((df['collegeGPA'] <= 10).sum())
```

#### `10percentage` and `12percentage` features are fine but `collegeGPA` feature needs to be perform some transformation

```python
# Create a boolean mask for rows where 'collegeGPA' is less than or equal to 10
mask = (df['collegeGPA'] <= 10)

# Multiply the values in 'collegeGPA' by 10 for rows satisfying the condition
df.loc[mask, 'collegeGPA'] *= 10
```

### 4. Checking 0's and -1's present or not in the dataset

```python
(df == 0).sum()[(df == 0).sum() > 0]
```

```python
(df == '0').sum()[(df == '0').sum() > 0]
```

#### Features like 10board, 12board and GraduationYear contains 0, we need to replace it with appropriate values

```python
# Replacing the 0 with mode in GraduationYear feature
df['GraduationYear'] = df['GraduationYear'].replace({0:np.nan})
df['GraduationYear'].fillna(df['GraduationYear'].mode()[0], inplace=True)
```

```python
# Replacing the 0 with mode in 10board feature
df['10board'] = df['10board'].replace({'0':np.nan})
df['10board'].fillna(df['10board'].mode()[0], inplace=True)
```

```python
# Replacing the 0 with mode in 12board feature
df['12board'] = df['12board'].replace({'0':np.nan})
df['12board'].fillna(df['12board'].mode()[0], inplace=True)
```

```python
(df == -1).sum()[(df == -1).sum() > 0] / df.shape[0]*100
```

```python
(df == '-1').sum()[(df == '-1').sum() > 0] / df.shape[0]*100
```

#### We're excluding the features like MechanicalEngg, ElectricalEngg, TelecomEngg and CivilEngg as these are optional subject columns with more than 80% missing values and filling the rest with zeros for analysis.

```python
df = df.drop(columns = ['MechanicalEngg', 'ElectricalEngg', 'TelecomEngg', 'CivilEngg'])
```

```python
# Replacing -1 with 0
df['ElectronicsAndSemicon'] = df['ElectronicsAndSemicon'].replace({-1:0})
df['ComputerScience'] = df['ComputerScience'].replace({-1:0})
```

```python
# Imputing Domain feature with the Median value
df['Domain'] = df['Domain'].replace({-1:np.nan})
df['Domain'].fillna(df['Domain'].median(), inplace = True)
```

```python
df['ComputerProgramming'] = df['ComputerProgramming'].replace({-1:np.nan})
df['ComputerProgramming'].fillna(df['ComputerProgramming'].median(), inplace = True)
```

```python
# Imputing cat feature like JobCity with the Mode 
df['JobCity'] = df['JobCity'].replace({'-1':np.nan})
df['JobCity'].fillna(df['JobCity'].mode()[0], inplace = True)
```

```python
df['Designation'].value_counts()
```

```python
df['Designation'].replace('systems engineer', 'system engineer', inplace=True)
```

### 5. Data Cleaning of text features

```python
text_feature = ['Designation','JobCity','10board','12board','Specialization','CollegeState']

for feature in text_feature:
    print('Number of unique values in {} feature: {}'.format(feature, df[feature].nunique()))
```

```python
df['JobCity'] = df['JobCity'].str.lower().str.strip()
df['JobCity'].nunique()
```

```python
from fuzzywuzzy import process

def jobcity(target_word = "", choices=[], threshold=80):
    match, score = process.extractOne(target_word, choices)
    if score >= threshold:
        return match
    else:
        return target_word
```

```python
choices = ['bangalore', 'indore', 'chennai', 'gurgaon', 'hyderabad', 'noida', 'kolkata', 'mohali', 'pune', 'delhi', 'mumbai']
```

```python
df['JobCity'] = df['JobCity'].apply(lambda city: jobcity(str(city), choices))
df['JobCity'].nunique()
```

```python
from scipy.stats import probplot

def generate_plots(df, column):
    # Summary Plot
    plt.figure(figsize=(5, 4))
    sns.despine()
    df[column].describe()[1:].plot(alpha = 0.8, marker = 'D', markersize = 8)    
    plt.title(f'Summary Statistics for {column}')
    plt.xlabel('Statistical Measures')
    plt.tight_layout()
    plt.show()

    # Histogram, KDE, and Box Plot
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 2, 1)
    sns.kdeplot(x=column, data=df)
    plt.title(f'KDE for {column}')
    plt.xlabel(column)

    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[column], color='lightblue')
    plt.title(f'Box Plot for {column}')
    plt.xlabel(column)

    plt.tight_layout()
    plt.show()

    # QQ Plot
    plt.figure(figsize=(5, 4))
    plt.title(f'QQ Plot for {column}')
    probplot(df[column], plot=plt)
    plt.xlabel('Theoretical Quantiles')
    plt.ylabel('Sample Quantiles')
    plt.show()
```

## b)- Is there a relationship between gender and specialisation? (i.e. Does the preference of Specialisation depend on the Gender?)

```python
res = pd.crosstab(df['Specialization'], df['Gender'])
res
```

```python
from scipy.stats import chi2, chi2_contingency
chi2_contingency(res)
```

```python
chi2_test_stat = chi2_contingency(res)[0]
p_val = chi2_contingency(res)[1]
data = chi2_contingency(res)[2]
```

```python
confidence_level = 0.95
alpha = 1 - confidence_level
chi2_critical = chi2.ppf(1 - alpha, data)
chi2_critical
```

```python
if(chi2_test_stat > chi2_critical):
    print("There is not enough evidence to reject Null Hypothesis")
else:
    print("There is sufficient evidence to reject Null Hypothesis")
```

```python
if(p_val < alpha):
    print("There is not enough evidence to reject Null Hypothesis")
else:
    print("There is sufficient evidence to reject Null Hypothesis")
```

### Observation

- As the result of the second research question we see that there is a relationship between Gender and specialization.
- We test this claim through Chi-Square test and find the result that both the categorical variables are dependent on each other.
